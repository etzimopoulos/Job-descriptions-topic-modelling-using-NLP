### Transforming Job Descriptions to NLP docs using Spacy

#### Indeed Job Descriptions

```python
nlp_jd = []
for jd in i_jobs['Job Description']:
    nlp_jd.append(nlp(jd))
```

```python

sf.show_entities(nlp_jd[5])
```

#### Kaggle Job Descriptions

```python
len(k_jobs['Job Description'][:100])
```

```python
nlp_jd_k = []
for jd in k_jobs['Job Description'][:100]:
    nlp_jd_k.append(nlp(jd))
```

```python
for i in nlp_jd_k:
    sf.show_entities(i)
    print('\n\n')
```

```python
for i in nlp_jd_k:
    for sent in i.sents:
        docx = nlp(sent.text)
        if docx.ents:
            displacy.render(docx, style='ent', jupyter=True)
#            doc, style='ent', options=options
        else:
            print(docx.text)
```
